```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
X = np.array([0, 2, 3, 4])
y = np.array([1, 3, 6, 8])

b1, b0 = 62. / 35, 18. / 35
X_space = np.linspace(-1, 5, num=100).reshape(-1, 1)
y_space = (b1 * X_space + b0).reshape(-1, 1)
```

```python
plt.style.use('ggplot')
plt.plot(X, y, 'ro')
plt.plot(X_space, y_space)
plt.savefig('./p2c.png', format='png')
```
